**Ethereal AI Pitch**

Idea: I want to make an AI-powered dating app. This is inspired off of my black mirror episode where they pair people together. I want the name to be Ethereal AI. The way this dating app will work by creating "personality" vectors that will be scored against other matches. Potential matches will be asked questions based on "how likely" you are etc. and then the LLM will match based off of the numbers in the vectors. The LLM will go further than just understanding interests or passions, but more *intuitive* traits such as **curiousity** **adventure** and **values**. While Tinder and Hinge may give you a snapshot of someone's profile, Ethereal AI will dig deeper.
