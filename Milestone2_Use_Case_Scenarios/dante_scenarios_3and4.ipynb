## Scenario 3: Job Search Matching – “Ethereal AI CareerMatch”

### Short Description

**“Discovering aligned jobs and teams through a 5-minute conversational career interview.”**

### Narrative of Scenario

The user opens Ethereal AI and selects the **Job & Career Matching** mode. Instead of uploading a traditional résumé only, they enter basic info (role, experience level, location) and then have a **5-minute guided conversation** with the AI.

The LLM asks about:

* Work style (independent vs. collaborative)
* What kind of managers they thrive under
* What drains them at work
* Their career values (impact, money, flexibility, prestige, learning, etc.)
* Concrete past experiences (projects they enjoyed vs. hated)

Behind the scenes, the app:

* Extracts **skills, preferences, and work-style traits** from the conversation
* Builds a structured **“career fit vector”** for the user
* Matches this vector against a database of job postings, annotated with similar feature vectors
* Produces a ranked list of job recommendations and recommended teams/environments

Over time, as the user gives feedback on matches (e.g., “not interested,” “interviewed here and liked it”), the system refines their profile and improves recommendations.


### Step-by-Step Breakdown of the Interaction

**High-level steps (bullets)**

1. **Mode Selection**

   * User chooses: “Find jobs & teams that fit me.”
2. **Light Onboarding**

   * App asks for: current role / experience level, location, salary expectations (optional), work authorization.
3. **5-Minute AI Career Conversation**

   * LLM asks open-ended questions:

     * “Tell me about a project that made you excited to go to work.”
     * “Describe a time you clashed with a manager—what happened?”
     * “What does your ideal workday look like?”
   * User responds in free-form text or voice (transcribed).
4. **Background Processing**

   * LLM parses the conversation into:

     * Core skills (e.g., Python, stakeholder management, copywriting)
     * Work-style traits (e.g., prefers autonomy, hates micro-management)
     * Values (impact, pay, stability, creativity, etc.)
     * Soft attributes (risk tolerance, communication style, conflict style).
   * Matching engine creates/updates the user’s **Career Profile Vector**.
5. **Job Matching**

   * Matching Engine compares user vector with job vectors (role requirements + inferred team culture).
   * Generates a ranked list of jobs with explanation blurbs (“This team values experimentation and autonomy, which you indicated you prefer.”).
6. **AI-Guided Shortlist & Prep**

   * LLM helps user shortlist 3–5 best-fit positions.
   * Offers tailored interview prep tips based on user’s strengths/weaknesses.
7. **Feedback Loop**

   * User marks jobs as “good fit,” “meh,” or “not relevant.”
   * System adjusts weightings in the user’s profile for future matches.



### Data Description

**Data needed (input to LLM/app)**

* **Structured user inputs:**

  * Name (or pseudonym), age range (optional), location, years of experience, field/industry, salary expectations (optional), work authorization.
  * Provided via forms & dropdowns.
* **Conversational data:**

  * Transcripts of the 5-minute conversation (free text or voice-to-text).
  * Includes narratives of past jobs, preferred team dynamics, frustrations, ambitions.
  * Provided to the LLM as conversation history.
* **Job data:**

  * Job postings (title, description, requirements, location, pay, seniority, company).
  * Optional: employer-provided culture tags (“fast-paced,” “flat hierarchy,” “high collaboration,” etc.).
* **Annotation schema / models:**

  * Psychometric mappings (e.g., Big Five, work-motivation frameworks) used by internal models to map raw text → trait scores.

**Data created (output / internal state)**

* **Career Profile Vector** for the user:

  * Skill embeddings vector (technical & soft skills).
  * Work-style traits (independence, collaboration, preference for structure, etc.).
  * Value weights (e.g., pay vs. purpose vs. flexibility).
  * Risk tolerance & preferred company stage (startup vs. big corp).
* **Job Embedding / Job Vector Store:**

  * LLM-generated vectors summarizing each job’s demands, environment, and growth profile.
* **Match records:**

  * (UserID, JobID, MatchScore, Timestamp, Explanation).
* **Feedback data:**

  * User reactions: liked/disliked, applied, got interview, got offer.
  * Used for retraining or fine-tuning matching logic (not necessarily the base LLM).


### Evaluation

**How to test successful completion of the scenario**

* **Offline evaluation:**

  * Use historical data (user résumés + jobs they applied to and liked) to test whether Ethereal AI’s recommended jobs overlap with ground truth or outperform keyword/skills-only matching.
* **Online metrics:**

  * CTR on recommended jobs.
  * Conversion to “apply” or “save” actions.
  * Long-term: self-reported fit (“How well does this job fit you?”), retention at matched jobs (if integrated with partner platforms).
* **User satisfaction surveys:**

  * Ask: “Did this feel more accurate than other job platforms?”
  * Likert scores for perceived understanding (“The AI understood what I want from a job”).

**Complications / error cases**

* **Vague answers:**

  * User only replies with “idk” or “I just want a good job.”
  * Mitigation: LLM switches to more concrete questions (“Tell me about a job you *didn’t* like and why”).
* **Adversarial or trolling answers:**

  * User feeds absurd or contradictory info (“I want to be a brain surgeon tomorrow but hate studying”).
  * Mitigation: detect low-quality input; flag risk; ask clarifying questions or reduce confidence and suggest broader matches.
* **Bias & fairness issues:**

  * Risk of reinforcing discriminatory patterns (e.g., by geography, school, age).
  * Mitigation: fairness constraints in the matching logic; regular audits; no use of protected attributes for ranking.
* **Privacy concerns:**

  * Sensitive career details, bad experiences with previous employers, etc.
  * Mitigation: clear consent flows, local redaction of sensitive fields, strong data encryption, and user controls over deletion.

GOAL
You are Ethereal AI RoomSync, a conversational LLM designed to match incoming college freshmen with compatible roommates. Your goal is to collect structured personal data, identify compatibility patterns, and recommend a roommate match based on at least a 6/10 alignment threshold across lifestyle, habits, expectations, and social preferences. You must maintain supportive, clarifying dialogue and help students reflect on compatibility.

PERSONA
You are a warm, thoughtful, highly structured matchmaker. You understand that students entering college may be nervous, unsure of what they want, or overly optimistic. You guide them gently, ask one question at a time, and help them articulate their needs clearly so the algorithm can identify a strong match. You never rush the student or assume anything without asking.
NARRATIVE
The student enters Ethereal AI RoomSync for the first time. The AI introduces itself, explains that it will ask a series of lifestyle and expectation-related questions to build a profile, and then compare that profile to existing students in the database. Only once enough information is gathered does RoomSync run its matching algorithm and present a recommended roommate. RoomSync continues until:
A match meets the 6/10 compatibility threshold,
The student approves the match, or
The student decides to stop using the system.
Follow These Steps in Order

STEP 1: GATHER INFORMATION
You should do this:
Introduce yourself and explain your purpose (helping find an ideal roommate).
Ask whether they are seeking a roommate.
Ask each of the following questions one at a time, waiting for a response after each:
What is your name?
What are your daily living habits (noise level, tidiness, temperature preferences, etc.)?
What do you expect from a roommate?
What are your dealbreakers?
What does your sleep schedule generally look like?
What is your ideal social environment (quiet, moderate, lively, etc.)?
You should also:
Ask questions sequentially, using natural dialogue.
Explain why you are asking each question.
Infer interaction style (structured, spontaneous, social, private) — but not personality judgments — to improve matching.
Adapt the flow of questions based on what is revealed.
Don’t do this:
Don’t ask multiple questions at once.
Don’t assume personality traits.
Don’t move on before the user responds.
STEP 2: COMPARE RESPONSES & FIND A MATCH
You should do this:
Scan the database for overlapping habits, expectations, and preferences.
Explain to the user that you are running the roommate matching algorithm.
Present one best-fit candidate with:
Key similarities
Potential differences (for transparency)
Ask the student to:
Confirm whether the match seems good
Describe why or why not they would choose this person
Don’t do this:
Don’t stop providing support after presenting the match.

STEP 3: WRAP UP
You should do this:
If the student approves the match (or declines the system entirely), close the conversation warmly.
Affirm that you’re excited for their college experience and available to help in the future.

## Scenario 4: Incoming College Freshman Roommate Matching – “Ethereal AI RoomSync”

### Short Description

**“Pairing incoming freshmen roommates through a conversational deep-dive into habits, boundaries, and social energy.”**

### Narrative of Scenario

An incoming freshman logs into the university’s housing portal and is redirected to **Ethereal AI RoomSync**. Instead of filling an impersonal survey with boxes like “Early bird vs. night owl,” they enter a few basics and then speak with the AI for ~5 minutes.

The LLM asks about:

* Sleep schedule and noise tolerance
* Cleanliness habits and mess tolerance
* Study routines
* Social life: parties, guests, introversion/extroversion
* Boundaries: sharing items, quiet-time, substance preferences, etc.

Ethereal AI turns this conversation into a **Roommate Compatibility Profile**. It then searches the pool of incoming students to find best-fit roommate pairs or small groups (for suites), balancing compatibility with diversity.

The AI can also draft a **Roommate Agreement Starter** for matched roommates based on shared and differing preferences.


### Step-by-Step Breakdown of the Interaction

**High-level steps (bullets)**

1. **Housing Integration**

   * User logs into the university housing portal and chooses: “Help me find a compatible roommate with Ethereal AI.”
2. **Basic Intake**

   * User provides: major/intended major, dorm preference, gender preferences for roommate, and required accommodations (e.g., quiet housing, disability considerations).
3. **5-Minute Rooming Conversation**

   * LLM asks:

     * “Describe your ideal weekday in your dorm—from waking up to going to sleep.”
     * “How do you feel about guests or parties in your room?”
     * “Are you usually the messy one or the neat one at home?”
     * “What’s something that would absolutely drive you crazy in a roommate?”
   * User responds in free text / voice.
4. **Trait Extraction & Profile Building**

   * LLM transforms answers into structured traits:

     * Chronotype (early/late), noise tolerance, cleanliness level, social energy, conflict style, academic seriousness, boundaries (e.g., substances, guests, shared items).
   * Creates **Roommate Profile Vector**.
5. **Matching**

   * Matching Engine compares vectors of all incoming freshmen.
   * Applies rules (e.g., same dorm, compatible gender preferences, no conflicts with accommodations).
   * Finds top N match candidates with compatibility scores.
6. **User Review & Confirmation**

   * App shows 3–5 high-compatibility roommate candidates (with anonymized summaries first, then optional reveal):

     * “Prefers early mornings, very neat, doesn’t like loud late-night parties, values quiet sleep.”
   * User can “star” or “pass” on candidates.
7. **Final Pairing**

   * Housing system uses mutual interest + high compatibility to finalize matches.
   * Ethereal AI generates a **Roommate Agreement Starter Document** summarizing:

     * Shared expectations, potential pressure points, suggested talking points for their first conversation.


### Data Description

**Data needed (input to LLM/app)**

* **Structured intake / constraints:**

  * Gender preference for roommate, dorm preferences, accessibility needs, required quiet hours (if any), program/major, preferred room type (double/triple/suite).
  * Provided via secure forms in the housing portal.
* **Conversational data:**

  * Text/voice responses about lifestyle, habits, and boundaries.
  * Provided as a chat transcript to the LLM.
* **Institutional data:**

  * List of all eligible students with their constraints, special housing assignments, and must-not-match rules (e.g., two people with conflicting accommodations).

**Data created (output / internal state)**

* **Roommate Profile Vector** per student:

  * Cleanliness score, social energy score, party tolerance, guest policy preferences, quiet-hours needs, substance preferences (where allowed to store), study seriousness, etc.
  * Represented as a feature vector plus human-readable descriptors.
* **Compatibility matrix:**

  * Pairwise compatibility scores between all students or between groups filtered by constraints.
* **Match candidates:**

  * For each student: list of top-K potential roommates with compatibility scores and explanation text.
* **Roommate Agreement Starter:**

  * A semi-structured document capturing shared preferences, differences, and suggested “house rules.”


### Evaluation

**How to test successful completion**

* **Pilot with freshmen cohort:**

  * Randomly divide students:

    * Control group → standard housing survey.
    * Treatment group → Ethereal AI RoomSync.
  * Compare:

    * Roommate conflict rates (reported conflicts, room change requests).
    * Self-reported satisfaction: “I feel compatible with my roommate.”
* **Short-term UX metrics:**

  * Completion rate of the 5-min conversation (do people drop off?).
  * Perceived accuracy: “This profile describes me well.”
* **Post-semester follow-up:**

  * Surveys about how well their roommate match fit their habits and preferences.

**Complications / error cases**

* **Under-reporting or aspirational answers:**

  * Students may claim they’re “very neat” or “not loud” but reality differs.
  * Mitigation: ask scenario-based questions (“What does your desk look like during midterms?”) to catch inconsistencies; highlight that honest answers directly affect their comfort.
* **Adversarial / trolling inputs:**

  * Students may intentionally give chaotic responses (“I never sleep, I party 24/7, I don’t care about anything”).
  * Mitigation: detect extreme or nonsense patterns; flag for manual review or revert to more robust defaults / follow-up questions.
* **Privacy & sensitivity:**

  * Sensitive topics (substance use, mental health, disabilities).
  * Mitigation: only store what’s necessary; clearly separate fields that go to housing admins vs. fields only used for compatibility calculations; student consent and anonymization for peers.
* **Over-optimization:**

  * If everyone is matched with “people exactly like them,” dorms might become socially siloed.
  * Mitigation: include a diversity factor in matching logic (e.g., compatible on core conflict dimensions like sleep/noise, but varied on interests and majors).
