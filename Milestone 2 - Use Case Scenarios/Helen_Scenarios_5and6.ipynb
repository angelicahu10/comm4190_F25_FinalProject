**Scenario 5**
*Matching with my classmates via google form and Ethereal AI*

A user in Penn's F25 "Talking with AI" class receives the Google Form link created. Their goal is to answer the survey questions honestly to provide the data needed for the matching algorithm. They navigate through the questions and rate their preferences on the 1-4 scales for lifestyle (e.g., "Early Bird" vs. "Night Owl"), and rate their interests (e.g., "Video Games," "Hiking"). After submitting the form, their "interaction" for this phase is complete. They now wait for Ethereal AI to run the analysis and present the results.


**Step by Step Breakdown of Interaction**

1. User receives Google Form
2. User answers all questions that judge personality, preference, working style, lifestyle on a scale of 1-4
3. User submits google form
4. Data is uploaded via google form


**Data**

What Data is needed: name, age, preferences, ALL questions answered via the google form. Data can be further broken down into
    1. Strings: eg. "Alice" or "Coffee Date"
    2. Integers eg. "3" or "4" from the 1-4 ranked questions 

*Data Created*
    When a google form is submitted, google automatically code responses in one line that contains all of a single user's answers (includes both strings and integers). This output is in raw vector format. 


*Data Translation*



1. Data Preprocessing/Translation 

One-Hot Encoding: For multiple-choice questions (like "Ideal first date") the "Ideal_Date" column becomes Ideal_Date_Coffee, Ideal_Date_Museum, etc. If a student chose "Coffee Date," they get a 1 in that column and 0 in all the others.

Normalization:  use of a MinMaxScaler. all the answers rescale to be a decimal (float) between 0.0 and 1.0. 1 becomes 0.0, 2 becomes 0.33, 3 becomes 0.66, and 4 becomes 1.0.

After this stage, your data is a clean, all-number table. Each row is a "user vector" that represents a student's entire personality.

2. The AI Model 
This model's job is to compare every single student to every other student. It does this by representing each student's vector as an arrow in a high-dimensional space (one dimension for each question).

It then measures the angle between every pair of arrows:

A small angle means the arrows point in the same direction. This tells the AI the two students had a very similar pattern of answers. They get a high score (e.g., 0.95, a 95% match).

A large angle (near 90°) means the arrows point in completely different directions. This means the students had very different or unrelated answers. They get a low score (e.g., 0.10, a 10% match).


**Evaluation**
1. make questions required on Google Form, so there are no missing components
2. Sanity check. If we run the model and find out everyone is a 95% match with eachother it probably means the model is doing something wrong
3. Are our questions evaluative enough? Do they reach beyond the surface level?
4. If the input from the user is unexpected how do we program/prompt the AI to flag it so the calculation of angles is not misaligned?

Description: A user finishes watching a movie on a streaming service powered by "Ethereal AI."  Often, the User wants more movies similar to those they like, but they don't want to sift through thousands of titles. Now, if they rate the movie with an Ethereal AI-powered platform, they can return to the homepage, which now displays a "Recommended for You" row. The AI, having instantly processed the Users rating, has updated its recommendations.

Step by Step Breakdown of Interaction

1. User  watches and rates a new piece of content (e.g., "Blade Runner 2049: 5 stars").

2. User navigates back to the homepage.

3. The system sends the new rating to Ethereal AI to update "user profile vector."

4. The AI model instantly compares User's updated vector to the vectors of all other users to find "tastemates" (users with the highest similarity score).

5. The AI identifies content (movies) that these tastemates loved but that User has not yet watched.

6. The AI ranks these movies and sends the top 3-5 back to the app.

7. The user sees these new, highly relevant movies in their "Recommended for You" list.

It now suggests Dune and Arrival—movies Alex hasn't seen but which other users with similar tastes (i.e., those who also loved Blade Runner 2049) have rated very highly.

**Data**

What Data is needed: The AI needs the user ratings. 

Rows are all the platform's Users (e.g., Alex, Bob, Charlie). These are strings.

Columns are all the platform's Items (e.g., Movie A, Movie B, Movie C). These are strings. 

The cell value is the **Integer** rating (e.g., 1-5) that a specific user gave to a specific movie. Most of this matrix is empty (zeros or "nulls") because no user has seen every movie.

A "null" value in the cell (ie. Helen, "Interstellar") is permanently changed to an Integer (e.g., 4).

This single change updates Helen's entire "raw user vector" (the row representing all of Helen's ratings).


**Data Translation**
Vector Creation: Each user's "vector" is simply their row from the user-item matrix. (ie. Helen_Vector = [5, 1, 0, 4, 5, ...], where 0 means "unrated").

Normalization (Mean-Centering): Unlike Google Form, don't use MinMaxScaler, but use Mean-Centering. The system calculates Helen's average rating (e.g., 3.5 stars) and subtracts it from every movie they rated.

A 5-star rating becomes +1.5.

A 1-star rating becomes -2.5. Normalization adjusts for bias (users who rate everything high vs. users who rate everything low), so the model compares their taste relative to their own average.

Exact model as Google Form model: Cosine Similarity.

This model's job is to compare every single user to every other user. 

It then measures the angle between every pair of arrows:

A small angle means the arrows point in the same direction. This tells the AI the two users had a very similar pattern of ratings (they liked and disliked the same types of movies). They get a high score (ie. 0.95, a "tastemate").

A large angle (near 90°) means the arrows point in completely different directions. This means the two users have very different tastes. They get a low score (ie., 0.10).
