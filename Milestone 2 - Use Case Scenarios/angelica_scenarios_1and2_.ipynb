**Scenario 1: Matching mentor-mentee pairings for research**

The user will fill out a survey with their preferences that indicate their working style and research interests. There will be a spot for them to write a few sentences of why they are interested in a specific professor's field of research, and the LLM will scan the text to find buzzwords that match up with certain projects to create a good pairing. 

*Interaction:*
- user creates an account
- user fills out survey/questionnare (either as student interested in research or as a professor looking for mentees)
- algorithm within LLM finds matching key words (ex. user1 = "interest in healthcare policy," user2 = "looking for passionate student interested in shaping healthcare policy," buzzword = "healthcare policy"
- LLM outputs match to user1 and user2, providing contact information and a convinent messaging system

*Data Description:*
- need user data on personal preferences and personality types to be able to match user1 and user2 responses
- compile information and run data through the trained algorithm sequence - which needs to be coded
- pairing occurs thorugh questions such as 'tell me about your work life balance/preferenes'

*Evaluation:*
- successful pairing if the system detects that user1 and user2 have been messaging on the platform
- success if matching rate from data is greater than 80%
- complication to consider -> the algorithm won't work well unless we acquire many users, otherwise there won't be anyone to match
- if users are not able to find a match, can potentially redirect them to a different resource

**Scenario 2: Using the compatability app to find a local excerise buddy**

The user will answer multiple choice survey questions, a few short answer responses, and film a quick introductory video that captures their tonality to match them to other users who have similar active interests. People whose personalities and vibes seem to match who are also interested in trying out the same activities will be paired through the LLM. When they go to message their partner, the LLM will use stored memory data to come up with automated messages that match the users voice, saving them time to think of what to say.

*Interaction:*
- user fills out survey answering what sport/active activity they want to do (ex. run a marathon, pilates, puppy yoga, tennis), provide their availability, location, budget, etc.
- algorithm will match users responses to one another and find a match
- LLM will take the data and also assist in finding a match, but also use short answer responses to create automated responses the user can use when reaching out through the messaging feature

*Data Description:*
- user information from the survey about personal active preferences and choices
- short answer responses will evaluate tone and personality type, LLM will predict how they write text messages
- as a result of the messaging interactions, each automated suggested response will become better and better

*Evaluation:*
- the app will track how often a user presses on the suggested response vs. how often they delete it and write their own
- if automated response is used more than 50% of the time, LLM proves to be beneficial and helpful in communciation
- if user1 and user2 match and meet up in person to perform the excercise, come back to the app to continue messaging, shows adherance
- saftey concerns if user1 and user2 meet in person, potential catfishing/scamming
