SCENARIO 4 — ROOMMATE MATCHING IMPLEMENTATION
Ethereal AI HomeSync
Explanation for the Diagram
Start (A):
The interaction begins with an introduction by Ethereal AI HomeSync, explaining that the system helps users find compatible roommates based on lifestyle and living preferences.
Ask: Are you looking for a roommate or offering a room? (B):
The user is prompted to identify their role. Based on their response, the workflow splits into two branches.
Renter / Seeker Pathway
Gather Seeker Information (C1):
This path collects living-preference details from users searching for a roommate.
Questions (D1, E1, F1, G1):
The seeker is asked about:
Daily schedule (early bird vs night owl)
Cleanliness standards
Noise tolerance and guest preferences
Budget range and location flexibility
Collect Seeker Responses (H1):
All responses are collected and structured for compatibility analysis.
Host / Provider Pathway
Gather Host Information (C2):
This path collects information from users offering a shared living space.
Questions (D2, E2, F2, G2):
The host is asked about:
Existing household routines
Cleanliness and shared-space expectations
Guest and noise policies
Lease constraints or house rules
Collect Host Responses (H2):
Responses are gathered and standardized.
Converging Matching Phase
Combine Living Preferences (I):
Preferences from both pathways are merged into a shared compatibility profile.
Search for Compatible Roommates (J):
The system scans its database for roommate pairs with aligned schedules, cleanliness norms, and boundary expectations.
Match Found: Present Roommate Profile (K):
A proposed roommate match is presented with a transparent explanation.
User Satisfaction Check (L):
The user confirms whether the match meets their expectations.
Confirm or Refine (M, N):
If satisfied, the match is confirmed.
If not, feedback is used to adjust weights and preferences.
Feedback Loop (O):
Unsuccessful matches inform future recommendations.
Wrap Up (P):
The session concludes with guidance for next steps or further refinement.
End:
Roommate matching session completes.


```mermaid
flowchart TD
    A[Start: Introduction by Ethereal AI HomeSync]
    B{Are you seeking<br/>a roommate or offering a room?}

    A --> B

    %% Seeker Path
    B -->|Seeking Roommate| C1[Gather Seeker Information]
    C1 --> D1[Q1: Daily Schedule Preference]
    D1 --> E1[Q2: Cleanliness Standards]
    E1 --> F1[Q3: Noise & Guest Tolerance]
    F1 --> G1[Q4: Budget & Location Flexibility]
    G1 --> H1[Collect Seeker Responses]

    %% Host Path
    B -->|Offering Room| C2[Gather Host Information]
    C2 --> D2[Q1: Household Routine]
    D2 --> E2[Q2: Cleanliness Expectations]
    E2 --> F2[Q3: Noise & Guest Policies]
    F2 --> G2[Q4: Lease Rules / Constraints]
    G2 --> H2[Collect Host Responses]

    %% Merge Paths
    H1 --> I[Combine Living Preferences]
    H2 --> I

    I --> J[Search Database for Compatible Roommates]
    J --> K[Match Found: Present Roommate Profile]

    K --> L{User satisfied<br/>with match?}
    L -->|Yes| M[Confirm Match & Next Steps]
    L -->|No| N[Refine Preferences]

    N --> J
    M --> O[Wrap Up & Support]


SCENARIO 5 — JOB MATCHING IMPLEMENTATION
Ethereal AI CareerSync
Explanation for the Diagram
Start (A):
The process begins with Ethereal AI CareerSync introducing itself as a career-alignment and job-matching tool.
Ask: Are you a job seeker or an employer role provider? (B):
The workflow branches depending on the user’s role.
Job Seeker Pathway
Gather Candidate Information (C1):
This pathway collects details from individuals seeking employment.
Questions (D1, E1, F1, G1):
The candidate is asked about:
Preferred work structure (structured vs flexible)
Schedule preferences
Skill interests and strengths
Risk tolerance and career priorities
Collect Candidate Responses (H1):
Responses are organized into a candidate profile.
Role / Employer Pathway
Gather Role Information (C2):
This pathway collects structured data for open job roles.
Questions (D2, E2, F2, G2):
The role provider specifies:
Expected work style and autonomy
Schedule and workload patterns
Required skills and responsibilities
Risk, growth, and stability profile
Collect Role Responses (H2):
Role attributes are collected and standardized.
Job Matching Phase
Combine Candidate and Role Profiles (I):
Candidate preferences and role requirements are merged.
Search for Compatible Job Matches (J):
CareerSync evaluates alignment across skills, work style, schedule, and risk tolerance.
Match Found: Present Role Fit (K):
A role is presented with a detailed compatibility explanation.
Candidate Satisfaction Check (L):
The candidate evaluates whether the role aligns with expectations.
Proceed or Refine (M, N):
If satisfied, next steps are recommended.
If dissatisfied, feedback adjusts future matches.
Feedback Loop (O):
Outcomes refine the matching model.
Wrap Up (P):
The interaction closes with career guidance and option expansion.
End:
Job matching session concludes.


```mermaid
flowchart TD
    A[Start: Introduction by Ethereal AI HomeSync]
    B{Are you a job seeker<br/>or role provider?}

    A --> B

    %% Job Seeker Path
    B -->|Job Seeker| C1[Gather Candidate Information]
    C1 --> D1[Q1: Preferred Work Structure]
    D1 --> E1[Q2: Schedule Preference]
    E1 --> F1[Q3: Skills and Experience]
    F1 --> G1[Q4: Career Goals and Risk Tolerance]
    G1 --> H1[Collect Candidate Responses]

    %% Role Provider Path
    B -->|Role Provider| C2[Gather Role Information]
    C2 --> D2[Q1: Role Structure]
    D2 --> E2[Q2: Schedule and Workload]
    E2 --> F2[Q3: Required Skills]
    F2 --> G2[Q4: Growth and Risk Profile]
    G2 --> H2[Collect Role Responses]

    %% Merge Paths
    H1 --> I[Combine Candidate and Role Profiles]
    H2 --> I

    I --> J[Search Database for Compatible Job Matches]
    J --> K[Match Found: Present Job Fit]

    K --> L{User satisfied<br/>with match?}
    L -->|Yes| M[Proceed with Next Steps]
    L -->|No| N[Refine Preferences]

    N --> J
    M --> O[Wrap Up and Career Guidance]
